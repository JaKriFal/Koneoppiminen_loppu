# 1.Tehtävä

```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
chronic_kidney_disease = fetch_ucirepo(id=336) 
  
# data (as pandas dataframes) 
X = chronic_kidney_disease.data.features 
y = chronic_kidney_disease.data.targets 
  
# metadata 
print(chronic_kidney_disease.metadata) 
  
# variable information 
print(chronic_kidney_disease.variables) 
```

## 1.1. Check the composition of the first and second datasets

```python
df_x = pd.DataFrame(X)
df_x
```

```python
df_y = pd.DataFrame(y)
df_y
```

## 1.2. Merge two datasets

```python
df_combined = pd.concat([df_x, df_y], axis=1)
df_combined
```

## 1.3. Remove unnecessary columns and rename target columns

```python
df_prep = df_combined.drop(columns=['rbc','pc','pcc','ba','sc','htn','dm','cad','appet','pe','ane'])
df_prep = df_prep.rename(columns={'bp': 'blood pressure',
                             'sg': 'specific gravity',
                             'al': 'albumin',
                             'su': 'sugar',
                             'bgr': 'blood glucose random',
                             'bu': 'blood urea',
                             'sod': 'sodium',
                             'pot': 'potassium',
                             'hemo': 'hemoglobin',
                             'pcv': 'packed cell volume',
                             'wbcc': 'white blood cell count',
                             'rbcc': 'red blood cell count'})
```

```python
df_prep
```

## 1.4. Change hemoglobin units from g/dl to g/l

```python
df_prep['hemoglobin'] = df_prep['hemoglobin'].apply(lambda x: x*10)
df_prep
```

## 1.5. Replace the class values from "ckd" and "ckd/t" with "a", and the "notckd" value with "c"

```python
df_prep['class'] = df_prep['class'].replace({'ckd': 'a', 'ckd/t': 'a', 'notckd': 'c'})
df_prep
```

## 1.6. Delete rows with 3 or more missing values

```python
df_prep = df_prep.dropna(thresh=len(df_prep.columns) - 2)
print( "Nuber of rows after deleting: ", {df_prep.shape[0]})
df_prep
```

```python
df_prep.dtypes
```

## 1.7. Splitting the data frame into two data frames

```python
affected_df = df_prep[df_prep['class'] == 'a']
control_df = df_prep[df_prep['class'] == 'c']
affected_df
```

## 1.8. Histograms for each numerical column in "affected_df" and "control_df"

```python
import matplotlib.pyplot as plt
def plot_data(df):
    plt.figure(figsize=(20, 18))
    
    # Select only numeric columns, excluding columns with classes
    numerical_columns = df.select_dtypes(include=['int', 'float']).columns
    
    for i, col in enumerate(numerical_columns):
        plt.subplot(3, 3, i + 1)  # 3x3 grid for the plots
        df[col].plot(kind='hist', title=col, edgecolor='black')
        plt.xlabel(col)
        plt.ylabel('Frequency')
    
    plt.tight_layout()  # For a better display
    plt.subplots_adjust(hspace=300, wspace=1)
    plt.show()
```

```python
a_df = affected_df[affected_df['class'] == 'a'].drop(columns='class')
plot_data(a_df)
```

```python
c_df = control_df[control_df['class'] == 'c'].drop(columns='class')
plot_data(c_df)
```

## 1.9. Plotting a correlation matrix

```python
# calculate correlation matrix
corr_affected = affected_df.drop(columns=['class']).corr()
corr_affected
```

```python
# create heatmap with Matplotlib

plt.figure(figsize=(10, 10))
plt.imshow(corr_affected, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_affected)), corr_affected.columns, rotation=90)
plt.yticks(range(len(corr_affected)), corr_affected.columns)
plt.show()
```

```python
corr_control = control_df.drop(columns=['class']).corr()
corr_control
```

```python
# create heatmap with Matplotlib

plt.figure(figsize=(10, 10))
plt.imshow(corr_control, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_control)), corr_control.columns, rotation=90)
plt.yticks(range(len(corr_control)), corr_control.columns)
plt.show()
```
